```python
# parse input:

def parse_input():
    with open('text') as f:
        list_numbers = []
        for line in f.readlines():
            list_numbers.append(line.strip("\n"))
    return list_numbers

def make_num_list(list):
    new_list = []
    for sentence in list:
        new_char = []
        for char in sentence:
            if char.isnumeric():
                new_char.append(char)
        new_list.append(new_char)
    return new_list

def make_new_list(new_list):
    sum_lines=0
    for line in new_list:
        if len(line)>1:
            num = line[0]+line[-1] #first and last num
        else:
            num =line[0]+line[0]
        num = int(num)
        sum_lines+=num
    return sum_lines
```

```python
list = parse_input()
num_list = make_num_list(list)
new_list = make_new_list(num_list)
print(new_list)
```

```python
#Loop through the strings and replace numbers spelled out with numbers (strings)
def parse_input_2():
    with open('text2') as f:
        list_numbers = []
        for line in f.readlines():
            list_numbers.append(line.strip("\n"))
    return list_numbers
def replace_spelled_out(list):

    hashmap_word={
        "zero":"0",
        "one":"1",
        "two":"2",
        "three":"3",
        "four":"4",
        "five":"5",
        "six":"6",
        "seven":"7",
        "eight":"8",
        "nine":"9",
    }
    new_list = []
    for sentence in list:
        for key in hashmap_word.keys():
            if "eightwo" in sentence:
                sentence = sentence.replace('eightwo','82')
            elif "oneight" in sentence:
                sentence = sentence.replace('oneeight','18') 
            elif key in sentence:
                sentence = sentence.replace(key,hashmap_word[key])
                
        new_list.append(sentence)

    return new_list
```

```python
list_2 = parse_input_2()
```

```python
replaced_list = replace_spelled_out(list_2)
```

```python
def word_to_digit(word):
    """Convert worded numbers to digits."""
    mapping = {
        'nine': '9', 'eight': '8', 'seven': '7', 'six': '6', 
        'five': '5', 'four': '4', 'three': '3', 'two': '2', 'one': '1'
    }
    return mapping.get(word, '')

def process_line(line):
    """Process each line to find the calibration value."""
    # Replace worded numbers with digits, longer words first
    for word in sorted(word_to_digit('').keys(), key=len, reverse=True):
        line = line.replace(word, word_to_digit(word))

    # Find the first and last digit
    digits = [char for char in line if char.isdigit()]
    if len(digits) >= 2:
        return int(digits[0] + digits[-1])
    return 0

def calculate_total_from_file(file_path):
    """Calculate the sum of calibration values from a file."""
    total = 0
    with open(file_path, 'r') as file:
        for line in file:
            total += process_line(line.strip())
    return total

# Replace 'file_path' with the path to your file
file_path = "text"
print(calculate_total_from_file(file_path))
```
