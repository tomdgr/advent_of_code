```python
def get_sum_recursive(dict_of_folders, folder):
    sum = 0
    # check if key has a value:

    for file in dict_of_folders[folder]:
        if "dir" not in file:
            sum += int(file.split(" ")[0])
        else:
            sum += get_sum_recursive(dict_of_folders, file.split(" ")[1])
    return sum
```

```python

# dict of folder with sum of files in folder:
dict_of_sum = {}

for keys in dict_of_folders.keys():
    sum_key = get_sum_recursive(dict_of_folders, keys)
    print(keys)
    dict_of_sum[keys] = sum_key
```
